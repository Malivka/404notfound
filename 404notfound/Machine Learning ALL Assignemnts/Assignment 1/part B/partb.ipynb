```python
#Question 1
"""
(a) Create a 1D NumPy array with values 1 to 20. 
(b) Extract all prime numbers from it. 
(c) Compute the mean and variance of the extracted primes.
"""
import numpy as np

a = np.arange(1, 21)
primes = np.array([x for x in a if all(x % i for i in range(2, int(x**0.5)+1)) and x>1])
mean_val, var_val = primes.mean(), primes.var()
primes, mean_val, var_val
```

```python
#Question 2
"""
(a) Create a 4×4 NumPy array with numbers 1 to 16. 
(b) Extract the 2×2 bottom-left sub-matrix. 
(c) Compute the determinant of the sub-matrix. 
"""

import numpy as np 
seq_arr = np.arange(1, 17).reshape(4, 4)
sub_matrix = seq_arr[2:4, 0:2]
det_val = np.linalg.det(sub_matrix)
```

```python
#Question 3
"""Create a DataFrame with 5 students and marks in 3 subjects. 
(b) Add a column for total and average marks. 
(c) Identify the topper and print their name with average."""

df = pd.DataFrame({
    'Name': ['Alice','Bob','Charlie','David','Eva'],
    'Math':[85,76,90,65,92],
    'Physics':[88,69,95,70,89],
    'Chemistry':[91,72,85,68,94]
})
df['Total'] = df[['Math','Physics','Chemistry']].sum(axis=1)
df['Average'] = df[['Math','Physics','Chemistry']].mean(axis=1)
topper = df.loc[df['Average'].idxmax()]
df, topper['Name'], topper['Average']
```

```python
#Question 4
"""(a) Simulate 1000 coin tosses using NumPy (1=Head, 0=Tail). 
(b) Count frequency of heads and tails. 
(c) Estimate probability of heads. Is it close to 0.5? Why/Why not?"""

np.random.seed(42)
tosses = np.random.randint(0, 2, 1000)
heads = tosses.sum()
tails = 1000 - heads
prob_heads = heads / 1000
heads, tails, prob_heads
```

```python
#Question 6
"""(a) Create a 3×3 NumPy array with values 1 to 9. 
(b) Find its transpose and inverse. 
(c) Verify that A × A⁻¹ ≈ I. """

A = np.array([[1,2,3],[0,1,4],[5,6,0]])
A_T = A.T
A_inv = np.linalg.inv(A)
I_approx = A @ A_inv
A, A_T, A_inv, I_approx
```

```python
#Question 7
"""(a) Generate random daily temperatures (30 values, range 20–40°C). 
(b) Find the hottest and coldest day. 
(c) Compute mean, median, and standard deviation of temperatures. 
"""
np.random.seed(42)
temps = np.random.uniform(20, 40, 30)
hottest_day, coldest_day = temps.argmax()+1, temps.argmin()+1
mean_temp, median_temp, std_temp = temps.mean(), np.median(temps), temps.std()
temps, hottest_day, temps.max(), coldest_day, temps.min(), mean_temp, median_temp, std_temp
```

```python
#Question 8
"""(a) Create a Pandas Series of marks for 8 students. 
(b) Replace all marks below 40 with “Fail”. 
(c) Count how many students passed."""

marks = pd.Series([55,72,38,90,45,30,65,80],index=[f'Student{i+1}' for i in range(8)])
marks_mod = marks.apply(lambda x: 'Fail' if x<40 else x)
num_passed = (marks_mod != 'Fail').sum()
marks_mod, num_passed
```

```python
#Question 9
""" (a) Generate 500 random integers between 1 and 6 (simulate dice rolls). 
(b) Count how many times each face appears. 
(c) Compute relative frequencies and compare with theoretical 1/6. """

np.random.seed(42)
rolls = np.random.randint(1,7,500)
counts = np.bincount(rolls)[1:]
rel_freq = counts / 500
rolls[:20], counts, rel_freq, 1/6
```

```python
#Question 12
"""(a) Create a NumPy array with integers from 1 to 30. 
(b) Reshape it into a 5×6 matrix. 
(c) Extract all even numbers from the matrix and compute their average."""

import numpy as np
arr = np.arange(1, 31)
matrix = arr.reshape(5, 6)
even_numbers = matrix[matrix % 2 == 0]
average_even = even_numbers.mean()
even_numbers, average_even
```

```python
#Question 13
""" (a) Create a DataFrame of 6 students with columns: Name, Age, Marks. 
(b) Select only students who scored above the overall average marks. 
(c) Display names of students younger than 20 whose marks are above 60."""

import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Age': [19, 21, 18, 22, 20, 17],
    'Marks': [75, 65, 80, 55, 90, 62]
}
df = pd.DataFrame(data)
average_marks = df['Marks'].mean()
above_average = df[df['Marks'] > average_marks]
young_high_scorers = df[(df['Age'] < 20) & (df['Marks'] > 60)]['Name']
above_average, young_high_scorers
```
