### **Cell 1 ‚Äì Imports**

```python
import matplotlib.pyplot as plt
import numpy as np
```

---

### **Cell 2 ‚Äì Simple Line Plot**

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

---

### **Cell 3 ‚Äì Date vs Temperature Plot**

```python
date = ["25/12", "26/12", "27/12"]
temp = [8.5, 10.5, 6.8]

plt.plot(date, temp)
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Date wise Temperature")
plt.grid(True)
plt.yticks(temp)
plt.show()
```

---

### **Cell 4 ‚Äì Customized Line Plot**

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', label='sin(x)')
plt.title('Customized Line Plot')
plt.xlabel('X values')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

---

### **Cell 5 ‚Äì Multiple Plots (sin & cos)**

```python
y2 = np.cos(x)

plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Multiple Lines')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

---

### **Cell 6 ‚Äì Subplots**

```python
fig, axs = plt.subplots(2)

axs[0].plot(x, y, 'r')
axs[0].set_title('sin(x)')

axs[1].plot(x, y2, 'b')
axs[1].set_title('cos(x)')

plt.tight_layout()
plt.show()
```

---

### **Cell 7 ‚Äì Bar Plot**

```python
x_vals = ['A', 'B', 'C', 'D']
y_vals = [3, 7, 5, 4]

plt.bar(x_vals, y_vals)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

---

### **Cell 8 ‚Äì Histogram (Random Data)**

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

---

### **Cell 9 ‚Äì Multiple Histograms**

```python
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, bins=10, edgecolor='black')
plt.hist(x2, bins=10, edgecolor='black')
plt.hist(x3, bins=10, edgecolor='black')
plt.show()
```

---

### **Cell 10 ‚Äì Scatter Plot (Random)**

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

---

### **Cell 11 ‚Äì Scatter Plot (Seeded)**

```python
np.random.seed(42)

X = np.random.rand(50)
Y = np.random.rand(50)

plt.scatter(X, Y, marker='o', s=100, edgecolors='black', alpha=0.5)
plt.show()
```

---

### **Cell 12 ‚Äì Box Plot (Single Dataset)**

```python
np.random.seed(10)
d = np.random.normal(100, 20, 200)

fig = plt.figure(figsize=(10, 7))
plt.boxplot(d)
plt.show()
```

---

### **Cell 13 ‚Äì Box Plot (Multiple Datasets)**

```python
np.random.seed(10)

d_1 = np.random.normal(100, 10, 200)
d_2 = np.random.normal(90, 20, 200)
d_3 = np.random.normal(80, 30, 200)
d_4 = np.random.normal(70, 40, 200)

d = [d_1, d_2, d_3, d_4]

fig = plt.figure(figsize=(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
ax.boxplot(d)
plt.show()
```
