## ðŸ”¹ 1. Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([15, 23, 18, 23, 24, 22, 22, 19, 19, 16, 24, 11, 24, 16, 23]).reshape(-1, 1)
Y = np.array([49, 63, 58, 60, 58, 61, 60, 63, 60, 52, 62, 30, 59, 49, 68])

# Fit model
lr = LinearRegression()
lr.fit(X, Y)

# Plot
plt.scatter(X, Y)
plt.plot(X, lr.predict(X))
plt.xlabel("Internal Marks")
plt.ylabel("External Marks")
plt.title("Linear Regression")
plt.show()

# Output
print("Intercept:", lr.intercept_)
print("Slope:", lr.coef_[0])
print("RÂ² Score:", lr.score(X, Y))
```

import numpy as np

## ðŸ”¹ 3. Polynomial Regression (Degree = 3)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([15, 23, 18, 23, 24, 22, 22, 19, 19, 16, 24, 11, 24, 16, 23]).reshape(-1, 1)
Y = np.array([49, 63, 58, 60, 58, 61, 60, 63, 60, 52, 62, 30, 59, 49, 68])

# Transform
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Fit model
pr = LinearRegression()
pr.fit(X_poly, Y)

# Plot
x_sorted = X[np.argsort(X.squeeze())]
y_sorted = pr.predict(X_poly)[np.argsort(X.squeeze())]

plt.scatter(X, Y)
plt.plot(x_sorted, y_sorted)
plt.xlabel("Internal Marks")
plt.ylabel("External Marks")
plt.title("Polynomial Regression (Degree = 3)")
plt.show()

# Output
print("Intercept:", pr.intercept_)
print("Polynomial Coefficients:", pr.coef_)
print("RÂ² Score:", pr.score(X_poly, Y))
```

## ðŸ”¹ 4. Logistic Regression (Pass / Fail)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Dataset
X = np.array([15, 23, 18, 23, 24, 22, 22, 19, 19, 16, 24, 11, 24, 16, 23]).reshape(-1, 1)
Y = np.array([49, 63, 58, 60, 58, 61, 60, 63, 60, 52, 62, 30, 59, 49, 68])

# Convert to binary class (Pass â‰¥ 57)
y_bin = (Y >= 57).astype(int)

# Fit model
logr = LogisticRegression()
logr.fit(X, y_bin)

# Plot sigmoid curve
x_range = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
plt.scatter(X, y_bin)
plt.plot(x_range, logr.predict_proba(x_range)[:, 1])
plt.xlabel("Internal Marks")
plt.ylabel("Probability of Pass")
plt.title("Logistic Regression (Pass â‰¥ 57)")
plt.show()

# Output
print("Accuracy:", logr.score(X, y_bin))
print("Predicted Probabilities:\n", logr.predict_proba(X))
```

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([15, 23, 18, 23, 24, 22, 22, 19, 19, 16, 24, 11, 24, 16, 23]).reshape(-1, 1)
Y = np.array([49, 63, 58, 60, 58, 61, 60, 63, 60, 52, 62, 30, 59, 49, 68])

# Fit model
lr = LinearRegression()
lr.fit(X, Y)

# Plot
plt.scatter(X, Y)
plt.plot(X, lr.predict(X))
plt.xlabel("Internal Marks")
plt.ylabel("External Marks")
plt.title("Linear Regression")
plt.show()

# Output
print("Intercept:", lr.intercept_)
print("Slope:", lr.coef_[0])
print("RÂ² Score:", lr.score(X, Y))
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([15, 23, 18, 23, 24, 22, 22, 19, 19, 16, 24, 11, 24, 16, 23]).reshape(-1, 1)
Y = np.array([49, 63, 58, 60, 58, 61, 60, 63, 60, 52, 62, 30, 59, 49, 68])

# Transform
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Fit model
pr = LinearRegression()
pr.fit(X_poly, Y)

# Plot
x_sorted = X[np.argsort(X.squeeze())]
y_sorted = pr.predict(X_poly)[np.argsort(X.squeeze())]

plt.scatter(X, Y)
plt.plot(x_sorted, y_sorted)
plt.xlabel("Internal Marks")
plt.ylabel("External Marks")
plt.title("Polynomial Regression (Degree = 3)")
plt.show()

# Output
print("Intercept:", pr.intercept_)
print("Polynomial Coefficients:", pr.coef_)
print("RÂ² Score:", pr.score(X_poly, Y))
````
